```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
# dataset repository path
url = "https://raw.githubusercontent.com/chaegyeong/KHUDA_ML_10th/3653bd3d9f415148f5a30568cda2f220e36017ec/KHUDA_ML10th_WEEK1/ML10th_WEEK1.csv"
```

```python
df = pd.read_csv(url)
df.head()
```

```python
### 건들지마세요
n=5
answer = [0] * 6
```

## Task 1
df의 feature중 Fare의 평균값(소수점 첫째자리까지)을 구하고, 해당 값을 answer[1]에 저장하세요.

```python
## Input Box

mean_fare = 0

mean_fare = df['Fare'].mean()
mean_fare = round(mean_fare,1)

print(mean_fare)
```

```python
answer[1] = mean_fare
```

## Task 2
'Age'의 결측치를 승객 나이의 중앙값으로 채운 후, 'Age' > 20 인 승객의 수를 구하고, 해당 값을 answer[2]에 저장하세요.

```python
## Input Box

ans = 0
df['Age'] = df['Age'].fillna(df['Age'].median())
ans = len(df[df['Age'] > 20])

print(ans)
```

```python
answer[2] = ans
```

## Task 3
feature 'Fare'와 'Age'에 대하여 StandardScaler를 적용하고,'Fare'의 최댓값을 구하여 소수점 둘째 자리까지 반올림한 뒤, 해당 값을 answer[3]에 저장하세요.


```python
## Input Box

ans = 0

features = ['Fare', 'Age']
X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

ans = round(X_scaled[:, 0].max(), 2)

answer[3] = ans
print(answer[3])
```

## Task 4
Task3의 스케일링 된 데이터를 독립변수로, feature 'Survived'를 종속변수로 설정한 뒤, 교차검증을 실행합니다.
5회 교차 검증을 실행 후, 검증 정확도들의 평균값을 계산하여 소수점 4자리까지 구합니다. 해당 값을 answer[4]에 저장하세요.


```python
## Input Box

ans = 0
dt_clf = DecisionTreeClassifier(random_state=42)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

## 이 아래부터 작성해주세요

X = X_scaled
y = df['Survived']

scores = cross_val_score(dt_clf, X, y, cv=skf)
ans = round(np.mean(scores), 4)


answer[4] = ans
print(answer[4])
```

## Task 5
레이블 인코딩을 진행해주세요. 리스트의 값들을 모두 더해, 해당 값을 answer[5] 에 저장하세요.



```python
## Input Box

ans = 0

items = ['TV', '세탁기', '전자레인지', '믹서', '전자레인지', '드라이기', '세탁기', '드라이기', '컴퓨터', '컴퓨터', '선풍기']

encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
ans = labels.sum()

answer[5] = ans
print(answer[5])
```

```python
### 정답 채점 코드
import hashlib

ANSWER_URL = "https://raw.githubusercontent.com/chaegyeong/KHUDA_ML_10th/5467d129145762cd9fcc97ef208660cbbe0fcbee/KHUDA_ML10th_WEEK1/answer.csv"

def md5 (s) : return hashlib.md5(s.encode("utf-8")).hexdigest()

ans_df = pd.read_csv(ANSWER_URL)
gt = {int(r["task"]): str(r["answer"]).strip().lower()
      for _, r in ans_df.iterrows()}

wrong = []

for i in range(1, (n+1)) :
    user_answer = "" if answer[i] is None else md5(str(answer[i]).strip().lower())
    if (user_answer != gt.get(i, "")) : wrong.append(i)

print("탈출!" if not wrong else f"틀린번호 : {wrong}")
```
